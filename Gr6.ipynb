# Exploring Global Electricity Access in Schools Through Data Visualization to Support SDG 4 and SDG 7

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gdp = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_46.csv", skiprows=4)
urban = pd.read_csv("API_SP.URB.TOTL.IN.ZS_DS2_en_csv_v2_249.csv", skiprows=4)
electricity = pd.read_csv("API_EG.ELC.ACCS.ZS_DS2_en_csv_v2_158.csv", skiprows=4)
education = pd.read_csv("API_SE.XPD.TOTL.GB.ZS_DS2_en_csv_v2_9902.csv", skiprows=4)
```

```python
gdp.head()
```

```python
urban.head()
```

```python
electricity.head()
```

```python
education.head()
```

```python
year = "2019"
```

```python
gdp = gdp[["Country Name", year]].rename(columns={year: "GDP_per_capita"})
urban = urban[["Country Name", year]].rename(columns={year: "Urban_population"})
electricity = electricity[["Country Name", year]].rename(columns={year: "Electricity_access"})
education = education[["Country Name", year]].rename(columns={year: "Education_expenditure"})
```

```python
gdp.head()
```

```python
df = gdp.merge(urban, on="Country Name", how="inner")
df = df.merge(electricity, on="Country Name", how="inner")
df = df.merge(education, on="Country Name", how="inner")
```

```python
df = df.rename(columns={"Country Name": "Country"})
```

```python
exclude = [
    "World", "High income", "Low income", "Lower middle income",
    "Upper middle income", "OECD members", "Euro area",
    "Europe & Central Asia", "East Asia & Pacific",
    "Latin America & Caribbean", "Middle East & North Africa",
    "North America", "South Asia", "Sub-Saharan Africa",
    "European Union", "Arab World", "Central Europe and the Baltics"
]

df = df[~df["Country"].isin(exclude)]
```

```python
df = df.dropna()
```

#Key Visualizations

```python
plt.figure(figsize=(8,5))
plt.scatter(df["GDP_per_capita"], df["Electricity_access"])
plt.xlabel("GDP per capita")
plt.ylabel("Electricity access")
plt.title("GDP per capita vs Electricity Access")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Urban_population"], df["Electricity_access"])
plt.xlabel("Urban population")
plt.ylabel("Electricity access")
plt.title("Urban Population vs Electricity Access")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Education_expenditure"], df["Electricity_access"])
plt.xlabel("Education expenditure")
plt.ylabel("Electricity access")
plt.title("Education Expenditure vs Electricity Access")
plt.show()
```
